# Data Description

This dataset contains data related to employees of an organization, collected by the HR division. This data is available at https://www.kaggle.com/datasets/rishikeshkonapure/hr-analytics-prediction/ .
It includes 35 columns and 1470 rows with employee details.

- **Age**: Age of the employee in years. Continuous.

- **Attrition**: Target variable indicating whether the employee left the company. Categorical.

- **BusinessTravel**: Frequency of business travels. Ordinal.

- **DailyRate**: Daily wage paid to the employee. Continuous.

- **Department**: Employee department. Categorical.

- **DistanceFromHome**: Distance from employee's home to the office (in km). Continuous.

- **Education**: Highest education completed by employee. Ordinal.

- **EducationField**: Field of education. Categorical.

- **EmployeeCount**: Employees with the same details. Always 1 in this dataset. Continuous.

- **EmployeeNumber**: Unique Employee number used to identify employees. Unique for every row. Continuous.

- **EnvironmentSatisfaction**: Satisfaction with workplace environment. Ranges from 1-4. Ordinal.

- **Gender**: Gender of Employee. Categorical.

- **JobInvolvement**: Employee Involvement with job. Ranges from 1-4. Ordinal.

- **JobLevel**: Level of employee's role at the company. Ranges from 1-5. Ordinal.

- **JobRole**: Employee's Role. Categorical.

- **JobSatisfaction**: Satisfaction with job. Rangfes from 1-4. Ordinal.

- **Marital Status**: Employee's relationship status. Categorical.

- **MonthlyIncome**: Employee's monthly income taken home (currency not given). Continuous.

- **MonthlyRate**:  Employee's monthly rate to  be paid (currency not given). Continuous.

- **NumCompaniesWorked**: Number of companies employee has worked for. Continuous.

- **Over18**: Is employee over the age of 18. Always True. Categorical.

- **OverTime**: Does employee work overtime?. Categorical.

- **PercentSalaryHike**: Salary hike given to employee. Continuous.

- **PerformanceRating**: Employee's Performance Rating. Ranges from 3-4. Ordinal.

- **RelationshipSatisfaction**: Satisfaction with workplace relationships. Ranges from 1-4. Ordinal.

- **StandardHours**: Hours of work in a week. Always 80.

- **StockOptionLevel**: Company stock option level. Ranges from 0-3. Ordinal.

- **TotalWorkingYears**: Total years employee has worked in the company.

- **TrainingTimesLastYear**: Hours of training received by the employee over the last year. Continuous.

- **WorkLifeBalance**: Work-life balance according to employees. Ranges from 1-4. Ordinal.

- **YearsAtCompany**: Total years employee has worked at the current company. Continuous.

- **YearsInCurrentRole**: Total years employee has been in the current role. Continuous.

- **YearsSinceLastPromotion**: Total years since the employee received their last promotion. Continuous.

- **YearsWithCurrManager**: Total years with the current manager. Continuous.

# Analysis

We notice that variables such as JobSatisfaction, MaritalStatus and more do affect attrition levels, while variables such as PerformanceRating and Gender don't.

It is obvious that there are relationships between various variables in this dataset and employee attrition, and creating a model to predict attrition using certain variables should be feasible.

# Methods and Plan

My question of interest is: 

**What factors influence employee attrition, and how can we predict whether an employee will leave the organization?**

### Model:

**Logistic regression** is well-suited for this analysis because the outcome variable, attrition, is binary (yes/no). This model will be used to predict the probability of an event (e.g., employee attrition) occurring, making it ideal for classification problems where the response is categorical. Additionally, logistic regression allows us to interpret the relationship between independent variables (e.g., job satisfaction, salary) and the likelihood of attrition through its coefficients, which provide odds ratios.

### Variable Selection

We can use Lasso regression to select the best variables for building the model. By adding an L1 penalty to the logistic regression loss function, Lasso shrinks less important coefficients to zero, effectively performing variable selection. We can then use metrics such as AIC to select a model with good-fit.

### Assumptions:

To apply logistic regression effectively, several assumptions must be considered:
1. The dependent variable must be binary: Attrition is binary.
2. Observations should be independent of one another: Employee data doesn't affect the data of other employees.
3. There should be a linear relationship between the log-odds of the dependent variable and the continuous predictors.
4. Multicollinearity among predictors should be minimal to avoid unreliable estimates: We remove variables that are highly collinear. 
5. Large sample size: The dataset size of 1470 should be large enough to predict data properly.

### Limitations

1. Assumes there should be a linear relationship between the log-odds of the dependent variable and the continuous predictors which might not exist.
2. Sensitive to multicollinearity so we have to remove highly collinear variables before creating the model.

# Implementation of a proposed model

```r
# We set it to a level to prevent prediction issues later

attrition_data$Attrition <- ifelse(attrition_data$Attrition == "Yes", 1, 0)
```

# Training/Test Data Split

```r
set.seed(123)

attrition_split <- initial_split(attrition_data, prop = 0.75, strata = Attrition)
attrition_selection <- training(attrition_split)
attrition_inference <- testing(attrition_split)
```

# Variable selection

We start creating a LASSO model to select the best variables for our model. 

```r
lasso_model <-
    cv.glmnet(select(attrition_selection,-Attrition) %>% as.matrix(), 
              attrition_selection$Attrition %>% as.matrix(),
              family = "binomial",
              alpha = 0.5)

lasso_model
```

```r
beta_lasso <- coef(lasso_model,s="lambda.min")

lasso_selected_covariates <- as_tibble(
        as.matrix(beta_lasso),
        rownames='covariate') %>%
        filter(covariate != '(Intercept)' & abs(s1) !=0) %>% 
        pull(covariate)

selected_vars <- attrition_selection |>
    select(all_of(lasso_selected_covariates), Attrition)

ols_model <- glm(formula = Attrition ~ .,
                    data = selected_vars,
                    family = binomial)
```

```r
# Predict probabilities
logistic_pred_prob <- predict(ols_model, newdata = attrition_inference, type = "response")

# Convert probabilities to binary outcomes (0 or 1) based on threshold (0.5)
logistic_pred <- ifelse(logistic_pred_prob > 0.5, 1, 0)


# # Model Evaluation: Calculate accuracy
logistic_accuracy <- mean(logistic_pred == attrition_inference$Attrition)

conf_matrix <- table(Predicted = logistic_pred, Actual = attrition_inference$Attrition)

logistic_accuracy
```

# Interpretation of Result

The model shows a high accuracy of 85.9%, but this is likely driven by the class imbalance, as the majority class (employees who do not leave) dominates the predictions. The high number of true negatives (non-leaving employees) skews the results, and further evaluation with metrics like recall and precision would better assess the model’s performance.
