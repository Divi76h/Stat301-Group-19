<font size = 5> **Methods and Planning** <font>

<font size = 4> **Question** <font>

In the wake of the <a href="https://www.weforum.org/stories/2021/11/what-is-the-great-resignation-and-what-can-we-learn-from-it/">Great Resignation</a> of 2021 and 2022, many companies have been scrambling to find ways to retain staff. During this period, companies have been offering more and more benefits to their employees, including work from home policies increasing wages and others benefits design to increase "Work-Life Balance'. The goal of this analysis is to explore the impacts of a work-life balance and benefits in that aim on the attrition rate of employees. We will use the variables `Attrition`, `DistanceFromHome`,`JobSatisfaction`,`WorkLifeBalance`,`EnvironmentSatisfaction`,`RelationshipSatisfaction` and `HourlyRate`. This is both an inference  and predictive analysis, as we are determining whether a link between the variables exist and predicting whether an employee attrits based off of the data.

<font size = 4> **Method** <font>

For our analysis, we will be using a logistic regression model. This model was selected because, as a categorical variable (i.e. Yes or No), logistic regression allows us to better model the data as compared to a linear model. A linear model would include a significant amount of plotting for in between variables, which is counterproductive for categorical variables (0.6 no is nonsensical and makes no real world sense). We will be comparing our model to a null model, essentially predicting the modal Attrition state. This null model will act as a baseline and will be used to analyze the effectiveness of the logistic regression model. 

<font size = 3> **Assumptions** <font>

The assumptions of the chosen logistic model include

- Linearity between the logit function and the independent variables
- Sample size must be sufficiently large (n=1470 is large enough)
- The variables aren't multicollinear
- The residuals are independent

<font size = 3> **Limitations** <font>

- The model assumes linearity between the logit and independent variables, which doesn't hold true in many real world cases.
- Incredibly sensitive to multicollinearity. Great care must be taken to ensure that the variables aren't colinear.

```r
raw_hr_data <-read_csv("https://raw.githubusercontent.com/Divi76h/Stat301-Group-19/refs/heads/main/data/HR-Employee-Attrition.csv")
head(raw_hr_data)
nrow(raw_hr_data)
colSums(is.na(raw_hr_data))
```

Before we start cleaning and wrangling the code, it can be seen that there are 1470 observations in the whole dataset, with no N/A values in any columns. Next, we can select our variables as outlined in the Methods section above (i.e, `Attrition`,`DistanceFromHome`,`JobSatisfaction`,`WorkLifeBalance`, and `HourlyRate`. The categorical variable of `Attrition` will be converted to a factor using the function `as_factor()`. 

```r
hr_data <- raw_hr_data |>
select(Attrition,DistanceFromHome,JobSatisfaction,WorkLifeBalance,HourlyRate,EnvironmentSatisfaction,RelationshipSatisfaction,)|>
mutate(Attrition = as_factor(Attrition)) |>
mutate(JobSatisfaction = as_factor(JobSatisfaction)) |>
mutate(WorkLifeBalance = as_factor(WorkLifeBalance)) |>
mutate(EnvironmentSatisfaction = as_factor(EnvironmentSatisfaction)) |>
mutate(RelationshipSatisfaction = as_factor(RelationshipSatisfaction)) 

head(hr_data)
```

```r

#Splitting Data into Training and Testing Sets

set.seed(123)
hr_split <- hr_data |>
initial_split(prop=0.7,strata="Attrition") 
hr_training <- training(hr_split)
hr_testing <- testing(hr_split)



baseline_pred <-tibble(response = "No",names = 1:nrow(hr_testing)) |>
mutate(response = as_factor(response))
baseline_pred


#baseline_accuracy <- mean(baseline_pred$x == hr_testing$Attrition)





logistic_model <- glm(formula = Attrition ~ .,
                      data = hr_training, 
                      family = binomial)



# Predict probabilities
logistic_pred_prob <- predict(logistic_model, newdata = hr_testing, type = "response") |>
tidy()


# Convert probabilities to binary outcomes (Yes or No) based on threshold (0.5)
logistic_pred <- logistic_pred_prob |>
mutate(x = ifelse(x < 0.5, "Yes", "No")) |> 
mutate(x=as_factor(x))


hr_training$Attrition
logistic_pred



# # Model Evaluation: Calculate accuracy
logistic_accuracy <- mean(logistic_pred$x == hr_testing$Attrition)

logistic_accuracy
#logistic_prediction_binary
#logistic_prediction_accuracy <- mean(logistic_prediction == hr_testing$Attrition)
#logistic_prediction_accuracy
```
